```python
import numpy as np
from scipy import constants as const
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import animation, rc
plt.style.use("seaborn-pastel")
import elementary
```

```python
part=elementary.Particle(0,0,1)
e=elementary.Electron(0,0,2)
pro=elementary.Proton(0,0,3)
```

```python
print(part.mass)
print(e.mass)
print(pro.mass)
```

```python
part.impulse(0,0,10*part.mass)
e.impulse(0,0,10*e.mass)
pro.impulse(0,0,10*pro.mass)
```

```python
xpoints = [i for i in np.arange(0, 5, 1e-2)]
ypoints = []
ypoints2 = []
ypoints3 = []
for i in range(500):
    part.impulse(0,0,-const.g*part.mass*.01)
    part.move(.01)
    ypoints.append(part.position[2])
for i in range(500):
    e.impulse(0,0,-const.g*e.mass*.01)
    e.move(1e-2)
    ypoints2.append(e.position[2])
for i in range(500):
    pro.impulse(0,0,-const.g*pro.mass*.01)
    pro.move(1e-2)
    ypoints3.append(pro.position[2])
# First create a figure, with "handle" stored in variable f
# The figsize is displayed in inches (on a printed page), with (width, height)
f = plt.figure(figsize=(8,6))
# Then create axes on the figure, with "handle" stored in variable a
a = plt.axes()

a.plot(xpoints, ypoints, label="Particle")
a.plot(xpoints, ypoints2, color="Red", label="Electron")
a.plot(xpoints, ypoints3, color="Green", label="Proton")

a.set(xlabel="Time (s)", ylabel="Position (m)", title="Particle Trajectory")
# Add a legend describing which curve is which
a.legend()
# Show the active plot to the screen
plt.show()
```

The particles all acted exactly as they would be expected to. There is minor separation at first because of a difference in starting position, then they all proceed to follow an identical curve as they began with identical momentum.
